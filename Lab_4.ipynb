**Ступенчатый график**.

В научных исследованиях используется для отображения изменения цифровых данных, когда каждое значение имеет **строго определенный уровень**.

Может использоваться для **отображения динамики показателей по временным промежуткам**.

Ступенчатый график строится с помощью функции **step()**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 7)
y = x
where_set = ['pre', 'post', 'mid'] # значения ставятся слева, справа или посередине интервала
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

for i, ax in enumerate(axs):
 ax.step(x, y, 'g-o', where=where_set[i]) # where - определяет место, где будет установлен шаг
 ax.grid()
```

**Стековый график**

Суть его в том, что графики отображаются друг над другом, и каждый следующий является суммой предыдущего и заданного.
Идея стек-графиков - показать «части целого» с течением времени.

Стековый график не подходит в тех случаях, когда используются и отрицательные значения свойств.

Для построения стекового графика используется функция **stackplot()**.


```python
# Номер дня
days = [1, 2,3, 4, 5]

# Количество часов день - Учеба
Studying = [7, 8, 6, 11, 7]

# Количество часов в день - Компьютерные игры
playing =  [8, 5, 7, 8, 13]
labels = ['Учеба', 'Игры']

# Построим Stackplot
plt.stackplot(days, Studying, playing, colors =['r', 'c'])

plt.xlabel('День')
plt.ylabel('Количество часов')
plt.title('Учеба vs Игры')
plt.legend(labels, loc='upper left')

plt.show()
```

**Stem-график**

Визуально **stem-график** выглядит как набор линий от точки с координатами (x, y) до базовой линии, в верхней точке которой ставится маркер.


```python
# Простой пример
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))

plt.stem(x, y)
plt.show()
```

```python
# Зададим параметры
x = np.arange(0, 10.5, 0.5)
y = np.array([(-0.2)*i**2+2*i for i in x])
plt.stem(x, y,
         linefmt='g--',    # тип и цвет вертикальной линии
         markerfmt='^',    # тип маркера
         bottom=1);        # y-координата базовой линии (по умолчанию 0)
```

**Диаграмма рссеяния (Scatter plot)**

**Диаграмма рассеяния**, является одной из самых простых и эффективных графических техник для изучения связи между двумя переменными. График позволяет получить визуальное представление о том, как связаны две переменные, и позволяет определить наличие или отсутствие зависимости между ними.

Для построения диаграммы рассеяния используется функция **scatter()**. В простейшем виде её можно получить, передав функции scatter() координаты x и y:


```python
x = np.arange(0, 10.5, 0.5)
y = np.cos(x)
plt.scatter(x, y);
```

Для более детальной настройки изображения необходимо воспользоваться дополнительными параметрами функции **scatter()**.

```python
import matplotlib.colors as mcolors

# Настроим цветовую палитру
# https://matplotlib.org/stable/gallery/color/named_colors.html
bc = mcolors.BASE_COLORS

# Зададим x и y
x = np.arange(0, 10.5, 0.25) # от 0 до 10 с шагом 0.25
y = np.sin(x)

# С помощью функции randint можно создать массив из целых чисел.
# Задаим массив из 42 (длина массива х) целых чисел, рандомно сгенерировав их от 1 до 8 (длина списка BASE_COLORS)
num_set = np.random.randint(1, len(mcolors.BASE_COLORS), len(x))
print(num_set)
print(len(mcolors.BASE_COLORS))
print(len(x))

# Зададим размер
sizes = num_set * 35
# и цвет для каждой точки на диаграмме рассеяния
colors = [list(bc.keys())[i] for i in num_set]

# римуем диаграмму рассеяния
plt.scatter(x, y, s=sizes, alpha=0.4, c=colors, linewidths=2, edgecolors='face')

# Рисуем кривую, чтобы "бусины были на ниточке"
plt.plot(x, y, 'g--', alpha=0.4);
```

**Диаграмма рассеяния** может быть использована в реальной жизни для многих целей.

Например, она может быть использована в анализе **медицинских** данных для изучения связи между уровнем холестерина и сердечно-сосудистым заболеванием, или в **бизнесе** - для изучения связи между рекламными затратами и продажами продукта.

Диаграмма рассеяния является мощным инструментом для **анализа данных**.

**Визуализация категориальных данных – столбчатые диаграммы**

**Категориальные данные** – это данные, которые принимают конечное число возможных значений. Например, если мы говорим о таком товаре как **фрукты**, у него могут быть следующие категориальные значения: **наименование**, **цвет**.

Для визуализации категориальных данных хорошо подходят **столбчатые диаграммы**. Для их построения используются функции:

• **bar()** – вертикальная столбчатая диаграмма;
• **barh()** – горизонтальная столбчатая диаграмма.

Построим простую гистограмму, оформим ее, используя знания, полученные на предыдущих Уроках:


```python
fig, ax = plt.subplots()

# задаем список - наименование фруктов
fruits = ['Яблоки', 'Черника', 'Вишня', 'Апельсины']
# задаем список - количество фруктов в кг
counts = [40, 100, 30, 55]
# определяем метки для столбцов
bar_labels = ['Красный', 'Синий', '_Красный', 'Оранжевый']
# задаем цвет для каждого столбца
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

# задаем параметры для диаграммы
ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_xlabel('Количество, кг')
ax.set_title('Фрукты: тип, количество и цвет')
ax.legend(title='Цвет фрукта')

plt.show()
```

Давайте возьмем немного **статистики**.

У функции **bar()** есть параметры **xerr**, **yerr** - величиены ошибок для графика.

Когда мы говорим о среднем значении - иногда необходимо иметь ввиду **доверительный интервал** для этого значения.

Допустим, у нас есть данные по продажам некоего товара за три недели:

```python
Week1 = [50, 55, 54, 49, 52, 48, 54]
Week2 = [58, 55, 60, 65, 57, 58, 55]
Week3 = [46, 49, 55, 60, 55, 51, 29]

# Найдем средние продажи в неделю:
Week1_mean = np.mean(Week1)
Week2_mean = np.mean(Week2)
Week3_mean = np.mean(Week3)

# Найдем стандартные отклонения (характеризует степень отклонения от среднего значения)
Week1_deviation = np.std(Week1)
Week2_deviation = np.std(Week2)
Week3_deviation = np.std(Week3)

# зададим высоту столбцов для столбчатой диаграммы по средним пордажам
bars = [Week1_mean, Week2_mean, Week3_mean]
bar_categories = ['Week1', 'Week2', 'Week3']
# зададим величину коридора для среднего значения (ошибки), равную стандартному отклонению
error_bars = [Week1_deviation, Week2_deviation, Week3_deviation]

plt.bar(bar_categories, bars, width=0.5, linewidth=2, color='g', edgecolor='y', yerr=error_bars)

plt.grid()
plt.show()
```

Этот график показывает, что размах по продажам на Неделе3 был больше, чем на 1 и 2 неделе - это мы видим и по данным - на Неделе 3 они более неоднородны (были продажи и 60 и 29).

```python
# Ошибку можно задавать и как скалярную величину - константу
x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(-20, 20, len(x))

plt.bar(x, y, width=0.5, linewidth=2, edgecolor='r', yerr=2, bottom=10)
plt.grid()
```

```python
# Мы можем задавать различные значения для параметров внешнего вида столбчатых диаграмм

np.random.seed(123)

groups = [f'P{i}' for i in range(7)]
counts = np.random.randint(0, len(bc), len(groups))
width = counts*0.1
colors = [['r', 'b', 'g'][int(np.random.randint(0, 3, 1))] for _ in counts]

plt.bar(groups, counts, width=width, alpha=0.6, bottom=2, color=colors, edgecolor='k', linewidth=2);
```

```python
index = np.arange(7)

bw = 0.3

plt.bar(index, Week1, bw, color='b')
plt.bar(index+bw, Week2, bw, color='g')
plt.bar(index+2*bw, Week3, bw, color='r')
plt.xticks(index+1.5*bw,['Пн','Вт','Ср', "Чт", "Пт", "Сб", "Вск"])

labels = ['Неделя1', 'Неделя2', 'Неделя3']

plt.title('Продажи по дням за три недели', fontsize=14)
plt.ylabel('Продажи, шт.')
plt.legend(labels, bbox_to_anchor=(1, 0.6)) #Вынесем легенду за поле графика

plt.show()
```

**Круговые диаграммы**

**Круговые диаграммы** – это наглядный способ показать доли компонентов в наборе. Они идеально подходят для отчётов, презентаций и т.п.

Для построения круговых диаграмм в Matplotlib используется функция **pie()**.


```python
# Построим диаграмму, отражающую процентное соотношение объемов продаж фруктов
vals = [40, 100, 30, 55]
labels = ['Яблоки', 'Черника', 'Вишня', 'Апельсины']
explode = (0.1, 0, 0.15, 0)     # список долей, выносимых из диаграммы - определяем - как далеко выносим

fig, ax = plt.subplots()
ax.pie(vals, labels=labels,
       autopct='%1.1f%%',       # формат числа величины доли внутри сегмента (у нас - один знак после запятой)
       shadow=True,             # с тенью
       explode=explode,         # список долей, выносимых из диаграммы
       rotatelabels=True )      # поворачиваем подписи
ax.axis('equal');               # равное соотношение (см. https://matplotlib.org/stable/gallery/subplots_axes_and_figures/axis_equal_demo.html)
```

Используя словарь **wedgeprops**, можно сформировать диаграмму с пустотой внутри:

```python
vals = [40, 100, 30, 55]
labels = ['Яблоки', 'Черника', 'Вишня', 'Апельсины']
explode = (0.1, 0, 0.15, 0)

fig, ax = plt.subplots()
ax.pie(vals, labels=labels,
       autopct='%1.1f%%',       # формат числа величины доли внутри сегмента (у нас - один знак после запятой)
       shadow=True,             # с тенью
       explode=explode,         # список долей, выносимых из диаграммы
       rotatelabels=True,       # поворачиваем подписи
       wedgeprops=dict(width=0.7, linestyle='--', edgecolor='k'))
ax.axis('equal');
```

Диаграмма показывает, что их всех фруктов Черники было продано больше всего (почти половина продаж, а вишни - меньше всего.
