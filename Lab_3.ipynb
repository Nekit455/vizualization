Давайте вспомним предыдущий Урок:

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]

# Построение простого графика по заданным массивам координат x и y
plt.plot(x,y)
```

```python
# Изменим наш график - добвим маркеры, изменим тип линии и цвет:
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
plt.plot(x, y, marker = '^', color = 'g', linestyle='dashdot')
```

**Цвет линии** графика задаётся через параметр **color** (или **c**, если использовать сокращённый вариант). Значение может быть представлено в одном из следующих форматов:

1)	RGB или RGBA: кортеж (или список) значений с плавающей точкой из трех элементов, задающих компоненты R, G и B соответственно.

2) Также цвет можно задавать в формате RGB с использованием символа "#", как это принято в HTML и CSS, задавая компоненты цвета в 16-ричной системе счисления.

3) Особый случай задания цвета – это выбор серого цвета различной степени яркости.

4)	Cимвол из набора: {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}.

5)	имя цвета из палитры X11/CSS4.

6)	цвет из палитры xkcd.

7)	цвет из набора Tableau Color (палитра T10).

Более полуню информацию о том, какие есть **типы линий**, **типы маркеров** и **цветовые палитры** вы можете прочитать в файле **Урок3.pdf** или изучить **официальную техническую локументацию**.



Если цвет задаётся с помощью символа из набора
{'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'},
то он может быть совмещён со стилем линии в рамках параметра fmt функции **plot()**.

Например: штриховая красная линия будет задаваться так: '--r', а штрихпунктирная зелёная так '-.g':


```python
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
plt.plot(x, y, '-.g');
```

```python
# Можно построить несколько графиков, цвет для каждого графика будет подоран автоматически
# здесь мы задали только тип линии

x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
plt.plot(x, y1, '-', x, y2, '--', x, y3, '-.', x, y4, ':');
```

```python
# Приведем несколько примеров

x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]

plt.figure(figsize=(12, 7))

# График first
plt.plot(x, y1,
         'o-r',           # маркер ('o'), цвет ('r') = красный
         alpha=0.7,       # прозрачность (alpha)=0.7
         label='first',
         lw=5,            # толщина линии (lw)=5
         mec='b',         # цвет границы маркера (mec='b')=синий
         mew=2,           # толщина границы маркера (mew)=2
         ms=10)           # размер маркера (ms)=10

# График second
plt.plot(x, y2,
         'v-.g',          # маркер ('v'), стиль линии ('-'),
         label='second',
         mec='r',         # цвет границы маркера (mec='r')=красный
         lw=2,            # толщина линии (lw)=2
         mew=2,           # толщина границы маркера (mew)=2
         ms=12)           # размер маркера (ms)=12

plt.legend()
plt.grid(True)
```

```python
# Еше один пример
import numpy as np
import math # для вычисления математических функций

x = np.linspace(0, 10, 50)

y1 = [math.cos(i) for i in x] # Косинус
y2 = [math.sin(i) for i in x] # Синус

plt.figure(figsize=(12, 7))
plt.plot(x, y1, marker='s', linestyle='dashed',  color='deeppink', alpha=0.7, label='Косинус', lw=3, mec='teal', mew=2, ms=10)
plt.plot(x, y2, marker='s', linestyle='solid',  color='gold', alpha=0.7, label='Синус', lw=3, mec='darksalmon', mew=2, ms=12)
plt.legend()
plt.grid(True)
```

Для заливки областей используется функция **fill_between()**.

```python
# Создадим набор данных для эксперимента:
import numpy as np
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

# Отобразим график с заливкой:
plt.plot(x, y, c = 'r')
plt.fill_between(x, y);
```

```python
# Изменим правила заливки
plt.plot(x, y, c = 'r')
plt.fill_between(x, y, where = (y > 0.75) | (y < -0.75))
```

Используя параметры y1 и y2, можно формировать более сложные решения. Заливка области между 0 и y при условии, что y >= 0

```python
plt.plot(x, y, c = 'r')
plt.fill_between(x, y, where = (y > 0));
```

```python
# Вариант двухцветной заливки
plt.plot(x, y, c = 'r')
plt.grid()
plt.fill_between(x, y, where=y>=0, color='g', alpha=0.3)
plt.fill_between(x, y, where=y<=0, color='r', alpha=0.3);
```

А теперь посмотрим, как можно настраивать **маркировку графиков**.
Создадим набор данных:

```python
import numpy as np

x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

plt.plot(x, y, marker='o', c='g');
```

Мы видим, что график построен на 500 точках и **маркеры сливаются**. В этом случае нужно **задать интервал отображения маркеров**, для этого используется параметр **markevery**:

```python
# Варианты "расстановки" маркеров

x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

m_ev_case = [None, 10, (100, 30), slice(100,400,15), [0, 100, 200, 300], [10, 50, 100]]
fig, ax = plt.subplots(2, 3, figsize=(10, 7))

axs = [ax[i, j] for i in range(2) for j in range(3)]
for i, case in enumerate(m_ev_case):
  axs[i].set_title(str(case))
  axs[i].plot(x, y, 'o', ls='-', ms=7, markevery=case)
```

Для того, чтобы **отобразить только часть графика**, которая отвечает определённому условию, используйте предварительное **маскирование данных** с помощью функции **masked_where()** из пакета **numpy**

```python
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

y_masked = np.ma.masked_where(y < -0.5, y) # Задаем условие

plt.ylim(-1, 1)
plt.plot(x, y_masked, linewidth=3) # Рисуем только ту часть, которая удовлетворяет условию
```
